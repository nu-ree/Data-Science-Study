### 3분위수, 2분위수로 count 등급 구분

```python
for i, v in df_train.iterrows():
    if v['count'] >= 284:
        df_train.loc[i,'count_cat'] = "high"
    elif v['count'] < 284 and v['count'] >= 145:
        df_train.loc[i,'count_cat'] = "mid"
    else:
        df_train.loc[i, 'count_cat'] ="low"
```

### 데이터 전처리 및 시각화

```python
# datatime 변수에서 시간에 대한 정보를 꺼냄
# 그러기 위해서는 저것을 시간에 대한 자료로 바꿔야함
df_train['datetime'] = pd.to_datetime(df_train['datetime'])
df_train['year'] = [x.year for x in df_train['datetime']]
df_train['month'] = [x.month for x in df_train['datetime']]
df_train['dayofweek'] = [x.dayofweek for x in df_train['datetime']]
df_train['hour'] = [x.hour for x in df_train['datetime']]
df_train.drop('datetime', axis=1, inplace=True)
```

```python
categoryVariableList = ["hour","dayofweek","month","year","season","weather","holiday","workingday"]
for var in categoryVariableList:
    df_train[var] = df_train[var].astype("category")
```

```python
"""
* weather
1: Clear, Few clouds, Partly cloudy, Partly cloudy 
2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist 
3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds 
4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog

* season
1: "Spring", 2 : "Summer", 3 : "Fall", 4 :"Winter" 
"""

fig, axes = plt.subplots(2,3,figsize=(15,10))
sns.countplot(hue='count_cat', x="season", data=df_train, ax=axes[0,0])
sns.countplot(hue='count_cat', x="holiday", data=df_train, ax=axes[0,1])
sns.countplot(hue='count_cat', x="weather", data=df_train, ax=axes[0,2] )
sns.countplot(hue='count_cat', x="year", data=df_train, ax=axes[1,0])
sns.countplot(hue='count_cat', x="month", data=df_train, ax=axes[1,1])
sns.countplot(hue='count_cat', x="hour", data=df_train, ax=axes[1,2])
```

```python
dummy_list = ['season', 'weather', 'year',
              'month','dayofweek','hour']
```

```python
dummies_train = {}
dummies_test = {}
for dummy in dummy_list:
    # dummies_train이 사전이기 때문에
    # 사전의 key로 각 변수의 이름을 만들고
    # pd.get_dummies 함수를 이용해 더미를 만들어줌
    # 그러나 구분이 되지 않기 때문에
    # 접두사를 붙여줌 prefix
    dummies_train[dummy] = pd.get_dummies(df_train[dummy],
                                    prefix=dummy)
```

```python
# 더미변수를 만들었기때문에 원래 변수를 지우고
# 더미변수들을 더해줌
df_train_dum = df_train.drop(dummy_list, axis=1)
```

```python
for dummy in dummy_list:
    df_train_dum = pd.concat([df_train_dum, dummies_train[dummy]],
                        axis=1)
```

## 분류 문제로 풀기

```python
def model_select_classifier(X, y, m):
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3)   
    if m == 'lda':
        lda = LinearDiscriminantAnalysis(solver="svd")
        y_pred = lda.fit(x_train, y_train).predict(x_test)
        result = cross_val_score(lda, x_train, y_train, cv=5)
        return result.mean()
    elif m == 'rf':    
        rf = RandomForestClassifier()
        y_pred = rf.fit(x_train, y_train).predict(x_test)
        result = cross_val_score(rf, x_train, y_train, cv=5)
        return result.mean()
    elif m == 'lr':
        logit = LogisticRegression()
        y_pred = logit.fit(x_train, y_train).predict(x_test)
        result = cross_val_score(logit, x_train, y_train, cv=5)
        return result.mean()
    elif m == 'nb':
        nb = GaussianNB()
        y_pred = nb.fit(x_train, y_train).predict(x_test)
        result = cross_val_score(nb, x_train, y_train, cv=5)
        return result.mean()
    elif m == 'gb':
        gb = GradientBoostingClassifier()
        y_pred = gb.fit(x_train, y_train).predict(x_test)
        result = cross_val_score(gb, x_train, y_train, cv=5)
        return result.mean()
```

```python
df_x = df_train.drop(['count','count_cat'], axis=1)
df_y = df_train['count_cat']

model_res_c = []
models = ['lda', 'rf', 'lr', 'nb', 'gb']
for m in models:
    res = model_select_classifier(df_x, df_y, m)
    model_res_c.append((m,res))
sorted(model_res_c, key=lambda x: x[1],reverse=True)
```

## 회귀 문제로 풀기

```python
def model_select_resgressor(X, y, m):
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3)   
    if m == 'lda':
        lda = LinearDiscriminantAnalysis(solver="svd")
        y_pred = lda.fit(x_train, y_train).predict(x_test)
        result = cross_val_score(lda, x_train, y_train, cv=5)
        return result.mean()
    elif m == 'rf':    
        rf = RandomForestRegressor()
        y_pred = rf.fit(x_train, y_train).predict(x_test)
        result = cross_val_score(rf, x_train, y_train, cv=5)
        return result.mean()
    elif m == 'lr':
        logit = LogisticRegression()
        y_pred = logit.fit(x_train, y_train).predict(x_test)
        result = cross_val_score(logit, x_train, y_train, cv=5)
        return result.mean()
    elif m == 'nb':
        nb = GaussianNB()
        y_pred = nb.fit(x_train, y_train).predict(x_test)
        result = cross_val_score(nb, x_train, y_train, cv=5)
        return result.mean()
    elif m == 'gb':
        gb = GradientBoostingRegressor()
        y_pred = gb.fit(x_train, y_train).predict(x_test)
        result = cross_val_score(gb, x_train, y_train, cv=5)
        return result.mean()
```

```python
df_x = df_train.drop(['count','count_cat'], axis=1)
df_y = df_train['count']

model_res_r = []
models = ['lda', 'rf', 'lr', 'nb', 'gb']
for m in models:
    res = model_select_resgressor(df_x, df_y, m)
    model_res_r.append((m,res))
sorted(model_res_r, key=lambda x: x[1],reverse=True)
```
